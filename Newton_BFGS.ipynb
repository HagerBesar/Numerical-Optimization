```python
def Gradient(x0, max_it,alpha , df ):
    x=x0
     
    for i in range(max_it):
        print(f"\n****************** Iteration {i} ********************")
        x = x -  (df(x) *alpha)
        print(x)
        grad = np.abs(df(x))
        print(grad)
        if grad < 0.1:
            break
        
     
    
    return x , grad ,i
 
```

```python
def Newton(x0, max_it, df ,ddf):
    x=x0
     
    for i in range(max_it):
        print(f"\n****************** Iteration {i} ********************")
        x=x-df(x)/ddf(x)
        print(x)
        gred=np.abs(df(x))
        print(gred)
        if gred < 0.01:
            break
        
     
    
    return x,gred,i
 
x_min,gred,i = Newton(40,400,Df,DDf)
```

```python
print(f"Newton Results with x0 =40, epsilon=0.1\nFound solution after {i+1} iterations.")
print("x_min =" , x_min)
print( 'Gradient = ', gred )
```

```python
# Newton's method without learning rate, x0=40, epsilon = 0.1
```

```python
def Newton_Rate(x0, max_it,rate, df ,ddf):
    x=x0
     
    for i in range(max_it):
        print(f"\n****************** Iteration {i} ********************")
        x=x- (df(x)/ddf(x))*rate
        print(x)
        gred=np.abs(df(x))
        print(gred)
        if gred < 0.1:
            break
        
     
    
    return x,gred,i

 
x_min , gred ,i= Newton_Rate(40,200,0.8,Df,DDf)
```

```python
print(f"Newton Results  with x0 =40, epsilon=0.1, lr=0.8\nFound solution after {i+1} iterations.")
print("x_min =" , x_min)
print( 'Gradient = ', gred )
```

```python
f_1 = lambda x: x**2  + y**2
Df_1 = lambda X: np.array([2*X[0],2*X[1]])
DDf_1 = lambda X: np.array([[2,0],
                         [0,2]])
```

```python
def Gradient_Multi(x0, max_it,alpha ,epsilon, df ):
    x=x0
     
    for i in range(max_it):
        print(f"\n****************** Iteration {i} ********************")
        x = x -  (df(x) *alpha)
        print("xy =  ",x)
        grad = np.abs(df(x))
        print("grad=",grad)
        if grad[0] < epsilon and grad[1] < epsilon:
            break
#         if grad.sum() < epsilon :
#             break
        
    print(f"\n\n****************** Training Report ********************")  
    print(f"Found solution after {i+1} iterations.")
    print("xy_min =" , x)
    print('Gradient=', grad )
     
    
    return x , grad ,i
 
```

```python
def Newton_Multi(x0, max_it ,alpha, epsilon, df ,ddf):
    x=x0
     
    for i in range(max_it):
        print(f"\n****************** Iteration {i} ********************")
        x = x - alpha*(np.linalg.inv(ddf(x)) @ df(x))
        print("xy = ",x)
        grad = np.abs(df(x))
        print("grad=",grad)
        if grad[0] < epsilon and grad[1] < epsilon:
            break
        
    print(f"\n\n****************** Training Report ********************")  
    print(f"Found solution after {i+1} iterations.")
    print("xy_min =" , x)
    print('Gradient=',grad )
    
    return x , grad ,i
 
x_min , gred ,i = Newton_Multi((10,10), 200 ,1, 0.001 , Df_1 , DDf_1)

```
